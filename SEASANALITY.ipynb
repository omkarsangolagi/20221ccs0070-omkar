```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01","2022-11-01","2022-12-01"]),
    "value":[10,20,40,24,45,60,70,45,60,80,55,100]
}
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)

plt.figure(figsize=(10,6))
plt.plot(df.index, df["value"],marker='X')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```
