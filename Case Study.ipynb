```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/content/Bank_dataset.xlsx"
df = pd.read_excel(file_path)
print(df.head())
print(df.info())
print(df.describe())
print(df)
```

```python
df['Total Number of Loans'].fillna(df['Total Number of Loans'].mean(), inplace=True)
df['Number of Bad Loans'].fillna(df['Number of Bad Loans'].mean(), inplace=True)
df['Number of Good Loans'].fillna(df['Number of Good Loans'].mean(), inplace=True)
print(df)
```

```python
df['Percentage of Bad Loans'].fillna((df['Number of Bad Loans'] / df['Total Number of Loans']) * 100, inplace=True)
df['Percentage of Good Loans'].fillna((df['Number of Good Loans'] / df['Total Number of Loans']) * 100, inplace=True)
print(df)
```

```python
plt.figure(figsize=(30, 16))
plt.subplot(2, 2, 1)
sns.barplot(x='Age Group', y='Total Number of Loans', data=df)
plt.title('Number of Loans')
plt.subplot(2, 2, 2)
sns.barplot(x='Age Group', y='Number of Bad Loans', data=df)
plt.title('Number of BaaLoans')
plt.subplot(2, 2, 3)
sns.barplot(x='Age Group', y='Number of Good Loans', data=df)
plt.title('Numbe oF Good loans')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age Group', y='Total Number of Loans', data=df)
plt.title('Total NumberLoans')
plt.xlabel('Age Group')
plt.ylabel('Total Number of Loans')
plt.grid(axis='y', linestyle='-', alpha=0.9)
plt.show()
```

```python
plt.subplot()
plt.hist(df['Total Number of Loans'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram Total Number of Loans')
plt.xlabel('Total Number of Loans')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Number of Bad Loans'], df['Number of Good Loans'], color='orange', alpha=0.5)
plt.title('Scatter Plot ')
plt.xlabel('Number f Bad Loans')
plt.ylabel('Number of Good oans')
plt.grid(True)
plt.show()
```
