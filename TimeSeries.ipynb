```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
data={
    "Data":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-04-01","2022-05-01","2023-01-01","2023-03-01"]),
    "Value":[10,15,22,28,35,45,85,95]
}
df=pd.DataFrame(data)
df.set_index("Data", inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"], marker='X')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Treand")
plt.grid(True)
plt.show()
```
