```python

import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data="Python's popularity has led and of has the development of a frameworks wide range of and  frameworks and tools that further its enhance its capabilities."
word_cloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)

plt.figure(figsize=(10,5))
plt.imshow(word_cloud)
plt.title('word cloud')
plt.show()

words=text_data.split()
words_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(words_counts.keys(),words_counts.values())
plt.title('Hsitogram of word frequencies')
plt.xlabel('words')
plt.xticks(rotation=90)
plt.show()
```
