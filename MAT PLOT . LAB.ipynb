```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[10,20,30,40]
mylabels = ["apples", "bananas", "grapes", "papayas"]
mycolors = ["orange", "black", "darkred", "hotpink"]
myexplode = [0.1,0.1,0.1,0.1]
plt.pie(x, labels=mylabels , colors=mycolors , explode=myexplode , shadow=True)
plt.show()
```

```python
x=[1,2,3,4]
y=[2,3,4,5]
colors=["g","r","black","hotpink"]
plt.bar(x,y,color=colors)
plt.xlabel("x axis", fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.title("Bar plot")
plt.show()
```

```python
x=[1,2,3,4]
y=[2,3,4,5]
colors=["g","r","black","hotpink"]
sizes=[200,400,600,800]
plt.scatter(x,y,color=colors,s=sizes,alpha=0.8,edgecolor="black",marker="*")
plt.xlabel("x axis", fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.title("scatter plot")
plt.show()
```

```python
x=[1,2,3,4]
y=[2,3,4,5]
colors=["g","r","black","hotpink"]
plt.barh(x,y,color=colors)
plt.xlabel("x axis", fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.title("Barh plot")
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.subplot(2,2,2)
plt.pie([2],colors="r")
plt.subplot(2,2,3)
y=[20,40,50,60]
plt.pie(y)
r=[2,3,4,5]
s=[4,5,6,7]
plt.subplot(2,2,4)
plt.bar(r,s)
plt.show()
```
